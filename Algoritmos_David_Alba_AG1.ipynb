##Torres de Hanoi con Divide y vencerás

**Problema**

Las Torres de Hanoi es un problema  de recursión que consiste en mover una pila de discos de un poste de origen a un poste destino, utilizando un poste auxiliar como soporte. La restricción es que nunca se puede colocar un disco más grande sobre uno más pequeño.

**Estrategia del algoritmo**

Se basa en dividir el problema en subproblemas más pequeño

- Mover los N-1 discos superiores desde el poste origen al poste auxiliar.
- Mover el disco más grande desde el poste origen al poste destino.
- Mover los N-1 discos desde el poste auxiliar al poste destino.

Recursivamente cada paso reduce el problema hasta llegar al caso base (N=1).

**Implementación**

La función Torres_Hanoi de la actividad guiada

- Recibe como entrada el número de discos N y los identificadores de los postes (1, 2 y 3).
- En cada llamada recursiva, calcula el poste auxiliar mediante la expresión 6 - desde - hasta.
- Imprime cada movimiento indicando de qué poste a qué poste se mueve el disco.

NOTE: Hemos añadido un contador acumulado de movimientos para verificar el número total de pasos realizados.

**Complejidad temporal**

La complejidad del algoritmo es la siguiente recurrencia:

𝑇(𝑁) = 2*𝑇(𝑁−1) + 1T(N) = 2*T(N−1) + 1, donde 𝑇(𝑁) es el número de llamadas recursivas que necesitamos para resolver el problema con N discos

Primer nivel de recursión es 𝑇(𝑁) = 2*𝑇(𝑁−1) + 1T(N) = 2*T(N−1) + 1

Sustituyendo 𝑇(𝑁−1) tenemos que 𝑇(𝑁) = 2*[2*𝑇(𝑁−2) + 1] + 1 = 4*𝑇(𝑁−2) + 2 + 1 T(N) = 4*T(N−2) + 3

Sustituyendo 𝑇(𝑁−2): tenemos que 𝑇(𝑁) = 4*[2*𝑇(𝑁−3) + 1] + 3 = 8*T(𝑁−3) + 4 + 3 = 8*𝑇(𝑁−3) + 7

Y aqui podemos ver que  Y aqui podemos ver que T(N)=2^N−1 * T(1) + (2^N−1  − 1) = T(N) =2^N − 1

La complejidad temporal es

O(2^N)

**Complejidad espacial**

El algoritmo utiliza recursión, por lo que la memoria máxima requerida corresponde a la profundidad máxima de la pila de llamadas. La profundidad máxima es:

𝑂(𝑁)

**Correcion**

La solución es correcta porque cada subproblema corresponde exactamente al mismo problema con N-1 disco, y el caso base (N=1) es correcto. La combinación de los pasos resuelve el problema completo por inducción.

**Optimabilidad**

El algoritmo es exacto y, por tanto, siempre encuentra la solucion mas optima.

**Problema**

La sucesión de Fibonacci es una serie numérica donde cada término es la suma de los dos términos anteriores.

- F(0) = 1
- F(1) = 1
- F(n) = F(n-1) + F(n-2)

**Estrategia del algoritmo**

El algoritmo implementado sigue una estrategia recursiva, dividiendo el problema en dos subproblemas más pequeños en cada llamada:

- Si N < 2, devuelve 1 (caso base).
- En otro caso, calcula recursivamente:

  F(N-1) + F(N-2)

Cada llamada genera dos nuevas llamadas hasta que se alcanzan los casos base.

**Implementación**

La función Fibonacci:

- Recibe como entrada un número entero N, que representa el índice del término a calcular.
- Si N es menor que 2, devuelve 1.
- En caso contrario, suma las dos llamadas recursivas con N-1 y N-2.

**Complejidad temporal**

La complejidad temporal de esta implementación es exponencial O(2^N).
Esto ocurre porque cada llamada recursiva se expande en dos sub-llamadas, creando un árbol binario de profundidad N.

**Complejidad espacial**

La complejidad espacial viene determinada por la profundidad  de la pila de llamadas, que es: O(N) ya que cada rama del árbol de llamadas crece linealmente hasta alcanzar el caso base.

**Corrección**

La solución es correcta porque respeta la definición inductiva de la sucesión de Fibonacci: Los casos base son correctos: F(0) = 1 y F(1) = 1.

**Completitud**

El algoritmo es completo, dado que siempre termina para cualquier N >= 0 (la recursión está bien acotada por los casos base).

**Optimabilidad**

La solución es correcta pero no óptima, debido a la repetición innecesaria de cálculos en los mismos subproblemas. Por ejemplo, F(3) se calcula múltiples veces. Este problema puede resolverse usando:
Memorización. Programación dinámica (resolución iterativa y almacenamiento de resultados). Esto reduce la complejidad temporal a O(N).

```
#Sucesión_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

##Devolución de cambio por técnica voraz

**Problema**

Devolver un importe N usando el menor número posible de monedas de denominaciones predefinidas.

**Estrategia del algoritmo**

Algoritmo voraz donde en cada paso selecciona el mayor número de monedas de la denominación más grande que quepa en el importe restante. Este enfoque es muy eficiente para ciertos sistemas monetarios, pero no garantiza siempre la solución óptima en sistemas arbitrarios.

**Implementación**

La función cambio_monedas:

- Inicializa una lista SOLUCION con ceros.
- Recorre las denominaciones ordenadas de mayor a menor.
- Para cada denominación, calcula cuántas monedas puede usar sin superar N.
- Acumula el valor utilizado y termina si se alcanza N.

Ejemplo de ejecución correcto:

- Sistema de monedas: [25,10,5,1]
- Cantidad: 15
- Solución: [0,1,1,0] = (1 de 10 y 1 de 5)

Ejemplo de fallo:

- Sistema: monedas: [10,7,1]
- Cantidad: 14
- Solución: [1,0,4] (1 de 10 y 4 de 1)
- Solución óptima: [0,2,0] (2 de 7)

**Complejidad temporal**

Como el algoritmo recorre todas las denominaciones una sola vez en la lista, entonces la complejidad es

𝑂(n), donde n es el número de denominaciones.

**Complejidad espacial**

La lista SOLUCION tiene longitud n, entonces

𝑂(n)

**Correcion**

El algoritmo siempre produce una solución válida (puede no ser óptima).

**Completitud**

Es completo en el sentido de que siempre encuentra alguna solución que suma el importe deseado (N), siempre que exista una combinación de monedas que lo permita.

**Optimibalidad**

Es óptimo si el sistema monetario es canónico, es decir, si el enfoque voraz siempre coincide con el mínimo número de monedas (por ejemplo, monedas de potencias de 2).

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION

# Ejemplo donde funciona
res1 = cambio_monedas(15, [25,10,5,1])
print("Cambio de monedas  para 15 con [25,10,5,1]:", res1)

# Ejemplo donde falla (monedas 10, 7, 1)
res2 = cambio_monedas(14, [10,7,1])
print("Cambio de monedas para 14 con [10,7,1]:", res2)
```

## N-Reinas por técnica de vueta atrás


**Problema**

Colocar N reinas en un tablero de N×N de manera que:
- Ninguna reina ataque a otra en la misma fila.
- Ninguna reina ataque a otra en la misma columna.
- Ninguna reina ataque a otra en la misma diagonal.

**Estrategia del algoritmo**

Se usa backtracking:

- Se colocan las reinas una a una por columnas (o etapas).
- En cada etapa, se prueban todas las posibles posiciones en filas.
- Si la configuración es prometedora (no hay conflictos), se continúa a la siguiente etapa.
- Si no es prometedora, se retrocede (backtrack).
- Cuando se colocan todas las reinas, se imprime la solución.

**Implementación**

- La función es_prometedora comprueba:
  - Que no hay dos reinas en la misma fila.
  - Que no hay dos reinas en la misma diagonal.

- La función reinas genera las configuraciones posibles y llama recursivamente.

- La función escribe dibuja el tablero.

**Complejidad temporal**

El algoritmo explora todas las configuraciones posibles de colocar N reinas en N columnas, probando en cada etapa todas las filas.

- Sin ninguna poda, en cada etapa puedes colocar la reina en N filas. Por tanto, el número de configuraciones posibles sería: N^N porque hay N posibilidades en cada una de las N columnas.

- Con poda (la function es_prometedora()), se descarta  muchas configuraciones inválidas (reinas en la misma fila o diagonal). En la primera columna puedes elegir cualquier fila (N opciones).En la segunda columna puedes elegir cualquier fila que no esté ocupada por la primera reina (N−1 opciones).En la tercera columna, cualquier fila distinta de las dos anteriores y que no esté en diago Así hasta N columnas. Por tanto, el número de configuraciones válidas como máximo está acotado por: N! porque no puede haber dos reinas en la misma fila. Por tanto, la complejidad temporal es O(N!)

**Complejidad espacial**

Cada llamada recursiva consume espacio en la pila de llamadas. La profundidad máxima de la recursión es N (una llamada por columna)  y la lista solucion  tiene tamaño N. Por tanto, la complejidad espacial es O(N)

**Completitud**

Es completo porque explora todas las configuraciones posibles dutilizando backtracking. Si existe al menos una solución, la encontrará. Si no hay solución, recorrerá todo el espacio de búsqueda y confirmará que no hay ninguna.

**Optimalidad**

No aplicable porque el problema de las N Reinas no requiere optimizar un criterio

```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #  escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

# Ejemplo más pequeño de 4 reinas
print("Ejemplo con 4 reinas:")
reinas(4)

# Ejemplo standar de 8 reinas
print("\nEjemplo con 8 reinas:")
reinas(8)
```

## Viaje por el rio. Programación dinámica

**Problema**

Se trata de determinar el coste mínimo y la ruta óptima entre dos puertos de un río, dados los costes de los tramos directos e indirectos.

**Estrategia del algoritmo**

- Utiliza programación dinámica para calcular de forma eficiente el coste mínimo entre todos los pares de nodos.
- Aplica el principio de optimalidad de Bellman: "La solución óptima de un problema contiene soluciones óptimas a sus subproblemas."

- Para cada par de nodos (i,j):
	- Calcula el coste directo TARIFAS[i][j].
	- Verifica si pasar por algún puerto intermedio k es más barato.
	- Almacena en PRECIOS[i][j] el coste mínimo y en RUTA[i][j] el nodo intermedio elegido.

**Implementación**

NOTA: Algunos comentarios adicionales han sido añadidos al codigo fuente

- La matriz PRECIOS contiene los costes mínimos.
- La matriz RUTA guarda por dónde se pasa para reconstruir la ruta óptima.
- La función calcular_ruta() utiliza recursividad para obtener la secuencia de puertos.

**Complejidad temporal**

El bucle principal tiene dos niveles de anidación (i y j), y un bucle interno sobre k, por tanto, 𝑂(𝑁^3), porque se verifica todos los caminos posibles con un nodo intermedio.

**Complejidad espacial**

Dos matrices de tamaño N×N, por tanto, 𝑂(𝑁^2)

**Completitud**

Es completo porque garantiza encontrar el coste mínimo para todos los pares de nodos.

**Optimalidad**

El algoritmo es exacto y, por tanto, siempre encuentra la solucion mas optima.

```
# Matriz de adyacencia donde TARIFAS[i][j] contiene:
# El coste directo de ir de puerto i a puerto j.
# El valor 999 significa que no hay conexión directa.

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
# Calcula la matriz de precios mínimos entre todos los pares
# de puertos (PRECIOS), y determinaa la ruta óptima (RUTA).

  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      # Cálculo de mínimo
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      #  Verificación de rutas intermedias
      for k in range(i, j):
        # Para cada nodo intermedio k, comprobamos si ir de i a k y luego de k a j es más barato.
        # Si es más barato, actualizamos el mínimo y la ruta.
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    # Si estamos en el mismo nodo origen y destino, no hay pasos.
    #print("Ir a :" + str(desde))
    return ""
  else:
    # Si no, buscamos el nodo intermedio (k) almacenado en RUTA y seguimos recursivamente.
    # Así, reconstruimos la secuencia de puertos.
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
