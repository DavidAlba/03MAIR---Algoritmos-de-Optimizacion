##Torres de Hanoi con Divide y vencer√°s

**Problema**

Las Torres de Hanoi es un problema  de recursi√≥n que consiste en mover una pila de discos de un poste de origen a un poste destino, utilizando un poste auxiliar como soporte. La restricci√≥n es que nunca se puede colocar un disco m√°s grande sobre uno m√°s peque√±o.

**Estrategia del algoritmo**

Se basa en dividir el problema en subproblemas m√°s peque√±o

- Mover los N-1 discos superiores desde el poste origen al poste auxiliar.
- Mover el disco m√°s grande desde el poste origen al poste destino.
- Mover los N-1 discos desde el poste auxiliar al poste destino.

Recursivamente cada paso reduce el problema hasta llegar al caso base (N=1).

**Implementaci√≥n**

La funci√≥n Torres_Hanoi de la actividad guiada

- Recibe como entrada el n√∫mero de discos N y los identificadores de los postes (1, 2 y 3).
- En cada llamada recursiva, calcula el poste auxiliar mediante la expresi√≥n 6 - desde - hasta.
- Imprime cada movimiento indicando de qu√© poste a qu√© poste se mueve el disco.

NOTE: Hemos a√±adido un contador acumulado de movimientos para verificar el n√∫mero total de pasos realizados.

**Complejidad temporal**

La complejidad del algoritmo es la siguiente recurrencia:

ùëá(ùëÅ) = 2*ùëá(ùëÅ‚àí1) + 1T(N) = 2*T(N‚àí1) + 1, donde ùëá(ùëÅ) es el n√∫mero de llamadas recursivas que necesitamos para resolver el problema con N discos

Primer nivel de recursi√≥n es ùëá(ùëÅ) = 2*ùëá(ùëÅ‚àí1) + 1T(N) = 2*T(N‚àí1) + 1

Sustituyendo ùëá(ùëÅ‚àí1) tenemos que ùëá(ùëÅ) = 2*[2*ùëá(ùëÅ‚àí2) + 1] + 1 = 4*ùëá(ùëÅ‚àí2) + 2 + 1 T(N) = 4*T(N‚àí2) + 3

Sustituyendo ùëá(ùëÅ‚àí2): tenemos que ùëá(ùëÅ) = 4*[2*ùëá(ùëÅ‚àí3) + 1] + 3 = 8*T(ùëÅ‚àí3) + 4 + 3 = 8*ùëá(ùëÅ‚àí3) + 7

Y aqui podemos ver que  Y aqui podemos ver que T(N)=2^N‚àí1 * T(1) + (2^N‚àí1  ‚àí 1) = T(N) =2^N ‚àí 1

La complejidad temporal es

O(2^N)

**Complejidad espacial**

El algoritmo utiliza recursi√≥n, por lo que la memoria m√°xima requerida corresponde a la profundidad m√°xima de la pila de llamadas. La profundidad m√°xima es:

ùëÇ(ùëÅ)

**Correcion**

La soluci√≥n es correcta porque cada subproblema corresponde exactamente al mismo problema con N-1 disco, y el caso base (N=1) es correcto. La combinaci√≥n de los pasos resuelve el problema completo por inducci√≥n.

**Optimabilidad**

El algoritmo es exacto y, por tanto, siempre encuentra la solucion mas optima.

```
contador_movimientos = [0]

def Torres_Hanoi(N, desde, hasta):
  if N ==1 :
    print("Lleva la ficha " ,desde , " hasta ", hasta )
    contador_movimientos[0] += 1

  else:
    #Torres_Hanoi(N-1, desde, 6-desde-hasta )
    Torres_Hanoi(N-1, desde, 6-desde-hasta )
    print("Lleva la ficha " ,desde , " hasta ", hasta )
    contador_movimientos[0] += 1
    #Torres_Hanoi(N-1,6-desde-hasta, hasta )
    Torres_Hanoi(N-1, 6-desde-hasta  , hasta )

# Ejemplo con 3 discos
contador_movimientos= [0]
print(f'Ejemplo con tres discos')
Torres_Hanoi(3, 1 , 3)
print(f'Total de movientos: {contador_movimientos}')

# Ejemplo con 4 discos
contador_movimientos= [0]
print(f'Ejemplo con tres discos')
Torres_Hanoi(4, 1 , 3)
print(f'Total de movientos: {contador_movimientos}')
```

**Problema**

La sucesi√≥n de Fibonacci es una serie num√©rica donde cada t√©rmino es la suma de los dos t√©rminos anteriores.

- F(0) = 1
- F(1) = 1
- F(n) = F(n-1) + F(n-2)

**Estrategia del algoritmo**

El algoritmo implementado sigue una estrategia recursiva, dividiendo el problema en dos subproblemas m√°s peque√±os en cada llamada:

- Si N < 2, devuelve 1 (caso base).
- En otro caso, calcula recursivamente:

  F(N-1) + F(N-2)

Cada llamada genera dos nuevas llamadas hasta que se alcanzan los casos base.

**Implementaci√≥n**

La funci√≥n Fibonacci:

- Recibe como entrada un n√∫mero entero N, que representa el √≠ndice del t√©rmino a calcular.
- Si N es menor que 2, devuelve 1.
- En caso contrario, suma las dos llamadas recursivas con N-1 y N-2.

**Complejidad temporal**

La complejidad temporal de esta implementaci√≥n es exponencial O(2^N).
Esto ocurre porque cada llamada recursiva se expande en dos sub-llamadas, creando un √°rbol binario de profundidad N.

**Complejidad espacial**

La complejidad espacial viene determinada por la profundidad  de la pila de llamadas, que es: O(N) ya que cada rama del √°rbol de llamadas crece linealmente hasta alcanzar el caso base.

**Correcci√≥n**

La soluci√≥n es correcta porque respeta la definici√≥n inductiva de la sucesi√≥n de Fibonacci: Los casos base son correctos: F(0) = 1 y F(1) = 1.

**Completitud**

El algoritmo es completo, dado que siempre termina para cualquier N >= 0 (la recursi√≥n est√° bien acotada por los casos base).

**Optimabilidad**

La soluci√≥n es correcta pero no √≥ptima, debido a la repetici√≥n innecesaria de c√°lculos en los mismos subproblemas. Por ejemplo, F(3) se calcula m√∫ltiples veces. Este problema puede resolverse usando:
Memorizaci√≥n. Programaci√≥n din√°mica (resoluci√≥n iterativa y almacenamiento de resultados). Esto reduce la complejidad temporal a O(N).

```
#Sucesi√≥n_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesi√≥n de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

##Devoluci√≥n de cambio por t√©cnica voraz

**Problema**

Devolver un importe N usando el menor n√∫mero posible de monedas de denominaciones predefinidas.

**Estrategia del algoritmo**

Algoritmo voraz donde en cada paso selecciona el mayor n√∫mero de monedas de la denominaci√≥n m√°s grande que quepa en el importe restante. Este enfoque es muy eficiente para ciertos sistemas monetarios, pero no garantiza siempre la soluci√≥n √≥ptima en sistemas arbitrarios.

**Implementaci√≥n**

La funci√≥n cambio_monedas:

- Inicializa una lista SOLUCION con ceros.
- Recorre las denominaciones ordenadas de mayor a menor.
- Para cada denominaci√≥n, calcula cu√°ntas monedas puede usar sin superar N.
- Acumula el valor utilizado y termina si se alcanza N.

Ejemplo de ejecuci√≥n correcto:

- Sistema de monedas: [25,10,5,1]
- Cantidad: 15
- Soluci√≥n: [0,1,1,0] = (1 de 10 y 1 de 5)

Ejemplo de fallo:

- Sistema: monedas: [10,7,1]
- Cantidad: 14
- Soluci√≥n: [1,0,4] (1 de 10 y 4 de 1)
- Soluci√≥n √≥ptima: [0,2,0] (2 de 7)

**Complejidad temporal**

Como el algoritmo recorre todas las denominaciones una sola vez en la lista, entonces la complejidad es

ùëÇ(n), donde n es el n√∫mero de denominaciones.

**Complejidad espacial**

La lista SOLUCION tiene longitud n, entonces

ùëÇ(n)

**Correcion**

El algoritmo siempre produce una soluci√≥n v√°lida (puede no ser √≥ptima).

**Completitud**

Es completo en el sentido de que siempre encuentra alguna soluci√≥n que suma el importe deseado (N), siempre que exista una combinaci√≥n de monedas que lo permita.

**Optimibalidad**

Es √≥ptimo si el sistema monetario es can√≥nico, es decir, si el enfoque voraz siempre coincide con el m√≠nimo n√∫mero de monedas (por ejemplo, monedas de potencias de 2).

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION

# Ejemplo donde funciona
res1 = cambio_monedas(15, [25,10,5,1])
print("Cambio de monedas  para 15 con [25,10,5,1]:", res1)

# Ejemplo donde falla (monedas 10, 7, 1)
res2 = cambio_monedas(14, [10,7,1])
print("Cambio de monedas para 14 con [10,7,1]:", res2)
```

## N-Reinas por t√©cnica de vueta atr√°s


**Problema**

Colocar N reinas en un tablero de N√óN de manera que:
- Ninguna reina ataque a otra en la misma fila.
- Ninguna reina ataque a otra en la misma columna.
- Ninguna reina ataque a otra en la misma diagonal.

**Estrategia del algoritmo**

Se usa backtracking:

- Se colocan las reinas una a una por columnas (o etapas).
- En cada etapa, se prueban todas las posibles posiciones en filas.
- Si la configuraci√≥n es prometedora (no hay conflictos), se contin√∫a a la siguiente etapa.
- Si no es prometedora, se retrocede (backtrack).
- Cuando se colocan todas las reinas, se imprime la soluci√≥n.

**Implementaci√≥n**

- La funci√≥n es_prometedora comprueba:
  - Que no hay dos reinas en la misma fila.
  - Que no hay dos reinas en la misma diagonal.

- La funci√≥n reinas genera las configuraciones posibles y llama recursivamente.

- La funci√≥n escribe dibuja el tablero.

**Complejidad temporal**

El algoritmo explora todas las configuraciones posibles de colocar N reinas en N columnas, probando en cada etapa todas las filas.

- Sin ninguna poda, en cada etapa puedes colocar la reina en N filas. Por tanto, el n√∫mero de configuraciones posibles ser√≠a: N^N porque hay N posibilidades en cada una de las N columnas.

- Con poda (la function es_prometedora()), se descarta  muchas configuraciones inv√°lidas (reinas en la misma fila o diagonal). En la primera columna puedes elegir cualquier fila (N opciones).En la segunda columna puedes elegir cualquier fila que no est√© ocupada por la primera reina (N‚àí1 opciones).En la tercera columna, cualquier fila distinta de las dos anteriores y que no est√© en diago As√≠ hasta N columnas. Por tanto, el n√∫mero de configuraciones v√°lidas como m√°ximo est√° acotado por: N! porque no puede haber dos reinas en la misma fila. Por tanto, la complejidad temporal es O(N!)

**Complejidad espacial**

Cada llamada recursiva consume espacio en la pila de llamadas. La profundidad m√°xima de la recursi√≥n es N (una llamada por columna)  y la lista solucion  tiene tama√±o N. Por tanto, la complejidad espacial es O(N)

**Completitud**

Es completo porque explora todas las configuraciones posibles dutilizando backtracking. Si existe al menos una soluci√≥n, la encontrar√°. Si no hay soluci√≥n, recorrer√° todo el espacio de b√∫squeda y confirmar√° que no hay ninguna.

**Optimalidad**

No aplicable porque el problema de las N Reinas no requiere optimizar un criterio

```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la soluci√≥n tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est√° " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #  escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

# Ejemplo m√°s peque√±o de 4 reinas
print("Ejemplo con 4 reinas:")
reinas(4)

# Ejemplo standar de 8 reinas
print("\nEjemplo con 8 reinas:")
reinas(8)
```

## Viaje por el rio. Programaci√≥n din√°mica

**Problema**

Se trata de determinar el coste m√≠nimo y la ruta √≥ptima entre dos puertos de un r√≠o, dados los costes de los tramos directos e indirectos.

**Estrategia del algoritmo**

- Utiliza programaci√≥n din√°mica para calcular de forma eficiente el coste m√≠nimo entre todos los pares de nodos.
- Aplica el principio de optimalidad de Bellman: "La soluci√≥n √≥ptima de un problema contiene soluciones √≥ptimas a sus subproblemas."

- Para cada par de nodos (i,j):
	- Calcula el coste directo TARIFAS[i][j].
	- Verifica si pasar por alg√∫n puerto intermedio k es m√°s barato.
	- Almacena en PRECIOS[i][j] el coste m√≠nimo y en RUTA[i][j] el nodo intermedio elegido.

**Implementaci√≥n**

NOTA: Algunos comentarios adicionales han sido a√±adidos al codigo fuente

- La matriz PRECIOS contiene los costes m√≠nimos.
- La matriz RUTA guarda por d√≥nde se pasa para reconstruir la ruta √≥ptima.
- La funci√≥n calcular_ruta() utiliza recursividad para obtener la secuencia de puertos.

**Complejidad temporal**

El bucle principal tiene dos niveles de anidaci√≥n (i y j), y un bucle interno sobre k, por tanto, ùëÇ(ùëÅ^3), porque se verifica todos los caminos posibles con un nodo intermedio.

**Complejidad espacial**

Dos matrices de tama√±o N√óN, por tanto, ùëÇ(ùëÅ^2)

**Completitud**

Es completo porque garantiza encontrar el coste m√≠nimo para todos los pares de nodos.

**Optimalidad**

El algoritmo es exacto y, por tanto, siempre encuentra la solucion mas optima.

```
# Matriz de adyacencia donde TARIFAS[i][j] contiene:
# El coste directo de ir de puerto i a puerto j.
# El valor 999 significa que no hay conexi√≥n directa.

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
# Calcula la matriz de precios m√≠nimos entre todos los pares
# de puertos (PRECIOS), y determinaa la ruta √≥ptima (RUTA).

  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicializaci√≥n de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      # C√°lculo de m√≠nimo
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      #  Verificaci√≥n de rutas intermedias
      for k in range(i, j):
        # Para cada nodo intermedio k, comprobamos si ir de i a k y luego de k a j es m√°s barato.
        # Si es m√°s barato, actualizamos el m√≠nimo y la ruta.
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    # Si estamos en el mismo nodo origen y destino, no hay pasos.
    #print("Ir a :" + str(desde))
    return ""
  else:
    # Si no, buscamos el nodo intermedio (k) almacenado en RUTA y seguimos recursivamente.
    # As√≠, reconstruimos la secuencia de puertos.
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
